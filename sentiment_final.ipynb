```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalMaxPooling1D, Conv1D, LSTM
from tensorflow.keras import utils
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
import os
from tensorflow.keras.callbacks import ModelCheckpoint
%matplotlib inline 
import pymorphy3
import nltk
from tensorflow.keras import utils
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
from collections import Counter
nltk.download('punkt')
nltk.download('stopwords')

train = pd.read_csv('/Users/nikitaelkhin/Desktop/Faster/train.csv')[0:100]

def preprocess(text, stop_words, punctuation_marks, morph):
    tokens = word_tokenize(text.lower())
    preprocessed_text = []
    for token in tokens:
        if token not in punctuation_marks:
            lemma = morph.parse(token)[0].normal_form
            if lemma not in stop_words:
                preprocessed_text.append(lemma)
    return preprocessed_text

punctuation_marks = ['!', ',', '(', ')', ':', '-', '?', '.', '..', '...']
stop_words = stopwords.words("russian")
morph = pymorphy3.MorphAnalyzer()

train['preprocessed_texts'] = train.apply(lambda row: preprocess(row['text'], punctuation_marks, stop_words, morph), axis=1)

words = Counter()
for txt in train['preprocessed_texts']:
    words.update(txt)
word_to_index = dict()
index_to_word = dict()
max_words = 10000
for i, word in enumerate(words.most_common(max_words - 2)):
    word_to_index[word[0]] = i + 2
    index_to_word[i + 2] = word[0]

def text_to_sequence(txt, word_to_index):
    seq = []
    for word in txt:
        index = word_to_index.get(word, 1) # 1 означает неизвестное слово
        # Неизвестные слова не добавляем в выходную последовательность
        if index != 1:
            seq.append(index)
    return seq

train['sequences'] = train.apply(lambda row: text_to_sequence(row['preprocessed_texts'], word_to_index), axis=1)

preprocessed_train_data = train['sequences']
answers = utils.to_categorical(train['sentiment'])

maxlen = 200
preprocessed_train_data = pad_sequences(preprocessed_train_data, maxlen=maxlen)


model_lstm = Sequential()
model_lstm.add(Embedding(max_words, 32))
model_lstm.add(LSTM(16))
model_lstm.add(Dense(3, activation='softmax'))

model_lstm.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

history_lstm = model_lstm.fit(preprocessed_train_data, 
                              answers, 
                              epochs=2,
                              batch_size=128,
                              validation_split=0.1)

plt.plot(history_cnn.history['accuracy'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history_cnn.history['val_accuracy'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
test = pd.read_csv('/Users/nikitaelkhin/Desktop/Faster/test.csv')
test['preprocessed_texts'] = test.apply(lambda row: preprocess(row['text'], punctuation_marks, stop_words, morph), axis=1)
test['sequences'] = test.apply(lambda row: text_to_sequence(row['preprocessed_texts'], word_to_index), axis=1)
preprocessed_test_data = test['sequences']
preprocessed_test_data = pad_sequences(preprocessed_test_data, maxlen=maxlen)
n = len(preprocessed_test_data)
results = []
for i in range(n):
    preprocessed_test_data_array = np.array([preprocessed_test_data[i]])
    prediction = model_lstm.predict(preprocessed_test_data_array)
    prediction = prediction.tolist()[0]
    results.append(prediction.index(max(prediction)))

df = pd.DataFrame({'ID': range(n), 'sentiment': results})
df.to_csv('example.csv', index=False, encoding='utf-8')
file_path = os.path.abspath('example.csv')
print(file_path)
```
